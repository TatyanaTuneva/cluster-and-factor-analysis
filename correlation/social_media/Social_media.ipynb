Поиск корреляции между использованиями различных соцсетей на территории России

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Загрузка данных

```python
data = pd.read_csv('social_media-RU-monthly-201801-201812.csv')
data = pd.concat([data, pd.read_csv('social_media-RU-monthly-201901-201912.csv')], ignore_index=True)
data = pd.concat([data, pd.read_csv('social_media-RU-monthly-202001-202012.csv')], ignore_index=True)
data = pd.concat([data, pd.read_csv('social_media-RU-monthly-202101-202112.csv')], ignore_index=True)
data = pd.concat([data, pd.read_csv('social_media-RU-monthly-202103-202203.csv')], ignore_index=True)
data
```

Просмотр корреляции

```python
corr = data.corr()
corr
```

Тепловая карта

```python
plt.figure(figsize = (20,20))
plt.imshow(corr,cmap='seismic',interpolation='none',vmin=-1,vmax=1)
plt.colorbar()
plt.xticks(range(len(corr)),corr.columns)
plt.yticks(range(len(corr)),corr.columns);
```
